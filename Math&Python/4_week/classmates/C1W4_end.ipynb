```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Выбрано Гамма-распределение - [Wikipedia](https://ru.wikipedia.org/wiki/Гамма-распределение)

```python
gamma_rv = sts.gamma(2, scale=2) #задаем гамма функцию с параметрами k = 2, θ = 2)
sample = gamma_rv.rvs(1000) #выборка из 1000 значений
plt.hist(sample, bins=20, normed=True, color='gray', label='sample histogramm') #рисуем гистограмму

x = np.linspace(0,20,1000) # задаем вектор значений x на отрезке от 0 до 20 для pdf
pdf = gamma_rv.pdf(x) # вычисляем плотность распределения Гаммы
plt.plot(x,pdf, color = 'red', label ='pdf') #рисуем график функции распределения
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
```

Для Гамма распределения:

Мат.ожиджание $E[x]=kθ=4$

Дисперсия $D[x]=kθ^2=8$

```python
# Распределение выборочных средних для n = 5, 10, 50

n = 5
sample_avg5 = []
for a in range(1000): 
    sample_avg5.append(gamma_rv.rvs(n).mean()) #наполняем список средними
x = np.linspace(0,10,1000)
plt.hist(sample_avg5, bins=20, normed=True, color ='gray', label ='sample avg histogramm') #рисуем гистограмму средних
normal_rv = sts.norm(4, np.sqrt(8./n)) #функция нормального распределения(4=среднее Гамма, 8 = дисперсия Гамма, 8/n - дисперсия нормального распределения)
pdf5 = normal_rv.pdf(x) #плотность норм распределения
plt.plot(x, pdf5, color = 'red', label ='normal theoretical pdf') #рисуем график функции плотности распределения 
plt.legend()
plt.title('sample size n = %d' %n)
plt.xlabel('x')
plt.ylabel('number of samples')

    
```
