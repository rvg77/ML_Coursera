```python
import numpy as np
import math
from matplotlib import pylab as plt
%matplotlib inline
```

```python
def f(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.)
```

```python
x = np.arange(1, 15.1, 0.1)
y = [f(z) for z in x]
plt.plot(x, y)
plt.show()
```

```python
def get_y(w, x):
    return sum([w[i] * (x ** i) for i in range(len(w))])

def apr(points_x):
    b = np.array([f(x) for x in points_x])
    a = []
    for x in points_x:
        a.append([x ** i for i in range(len(points_x))])
    w = np.linalg.solve(a, b)
    points_y = [get_y(w, x) for x in points_x]
    plt.plot(points_x, points_y)
    plt.show()
    return w
```

```python
apr([1., 15.])
apr([1., 8., 15.])
print(apr([1., 4., 10., 15.]))
```
